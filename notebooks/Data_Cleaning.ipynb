# Data Cleaning and Preprocessing

Raw datasets often contain inconsistencies such as missing values, duplicate records, or incompatible data types. 
Proper data preprocessing ensures that the dataset is reliable and suitable for exploratory analysis and machine learning modeling.

In this notebook, the telecom customer churn dataset is inspected, cleaned, and transformed into a structured format suitable for downstream analysis.

## Importing Required Libraries

The analysis begins by importing the core Python libraries used for data manipulation and numerical operations.

```python
import pandas as pd
import numpy as np
```

## Data Loading and Initial Inspection

The telecom churn dataset is loaded into a pandas DataFrame. 
Inspecting the first few rows helps verify that the dataset has been imported correctly and provides an overview of the available features.

```python
churn_df = pd.read_csv("churn_dataset.csv")
```

```python
churn_df.head()
```

## Dataset Dimensions

Understanding the size of the dataset is an important initial step. 

```python
churn_df.shape
```

## Dataset Structure and Data Types

Examining the structure of the dataset helps identify the data types associated with each feature and detect potential issues such as incorrect types or missing values.

```python
churn_df.info()
```

## Missing Value Analysis

Missing values can affect statistical analysis and machine learning model performance. 
This step checks whether any features contain null values that require handling through imputation or removal.

```python
churn_df.isnull().sum()
```

## Duplicate Record Detection

Duplicate observations may distort analysis and model training. 
Therefore, the dataset is inspected for duplicate rows to ensure data integrity.

```python
churn_df.duplicated().sum()
```

## Encoding Categorical Variables

Machine learning algorithms require numerical input. 
The telecom dataset contains categorical features such as "International plan" and "Voice mail plan". 
These variables are converted into binary numerical representations to make them suitable for modeling.

```python
churn_df["International plan"] = churn_df["International plan"].map({ "Yes":1 , "No":0})
churn_df["Voice mail plan"] = churn_df["Voice mail plan"].map({ "Yes":1 , "No":0})
churn_df["Churn"] = churn_df["Churn"].map({ True:1 , False:0 })
```

```python
churn_df.head()
```

## Feature Name Standardization

For consistency and ease of use during analysis and modeling, feature names are standardized by converting them to lowercase and replacing spaces with underscores, also the removing of whitespaces from the text columns are too necessary.

```python
churn_df.columns = churn_df.columns.str.lower().str.replace(" ", "_")
churn_df = churn_df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

churn_df.head()
```

## Saving the Cleaned Dataset

After preprocessing, the cleaned dataset is stored separately to ensure reproducibility. 
Subsequent analysis notebooks will use this processed dataset rather than the raw data.

```python
churn_df.to_csv("cleaned_churn_dataset.csv", index = False)
```
